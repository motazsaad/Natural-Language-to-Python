```python
from google.colab import drive
drive.mount('/content/drive')

input_path = 'drive/My Drive/Colab Notebooks/Master thises/Data Preprocessing/input/input.txt'
model_saved_fol_path = 'drive/My Drive/Colab Notebooks/Master thises/Data Modeling/pytorch/06_model/'
```

```python
dataset = open(input_path).read().strip().split('EOC\n')
pairs = [line.split(' ||| ') for line in dataset]
```

```python
dps = {'nl': [], 'code': []}
```

```python
for inp, targ in pairs:
  dps['nl'].append(inp)
  dps['code'].append(targ)
```

avg line of code

```python
sum_lines_of_code = 0
for code in dps['code']:
  sum_lines_of_code = sum_lines_of_code + code.count('\n')
sum_lines_of_code/len(dps['code'])
```

avg word count of code

```python
sum_word_count_of_code = 0
for code in dps['code']:
  sum_word_count_of_code = sum_word_count_of_code + len(code.split())
sum_word_count_of_code/len(dps['code'])
```

avg word count of nl

```python
sum_word_count_of_nl = 0
for nl in dps['nl']:
  sum_word_count_of_nl = sum_word_count_of_nl + len(nl.split())
sum_word_count_of_nl/len(dps['nl'])
```

avg char count of code

```python
sum_char_count_of_code = 0
for code in dps['code']:
  sum_char_count_of_code = sum_char_count_of_code + len(code)
sum_char_count_of_code/len(dps['code'])
```

avg char count of nl

```python
sum_char_count_of_nl = 0
for nl in dps['nl']:
  sum_char_count_of_nl = sum_char_count_of_nl + len(nl)
sum_char_count_of_nl/len(dps['nl'])
```
